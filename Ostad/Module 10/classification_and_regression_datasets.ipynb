## Classification Practice

* [Titanic Survival](https://www.kaggle.com/competitions/titanic)
  Predict whether a passenger survived the Titanic disaster.

* [Diabetes Prediction](https://www.kaggle.com/datasets/uciml/pima-indians-diabetes-database)
  Predict whether a patient has diabetes based on medical features.

* [Iris Species Classification](https://www.kaggle.com/datasets/uciml/iris)
  Classify iris flowers into three species using flower measurements.

* [Breast Cancer Detection](https://www.kaggle.com/datasets/uciml/breast-cancer-wisconsin-data)
  Detect whether a tumor is malignant or benign.

## Regression Practice

* [House Price Prediction](https://www.kaggle.com/datasets/yasserh/housing-prices-dataset)
  Predict house prices from property features.

* [Advanced House Price Prediction](https://www.kaggle.com/competitions/house-prices-advanced-regression-techniques/data)
  Advanced regression challenge for predicting house sale prices.

* [Medical Insurance Charge](https://www.kaggle.com/datasets/mirichoi0218/insurance/)
  Predict medical insurance costs from personal attributes.

* [Sales Prediction from Advertising Cost](https://www.kaggle.com/datasets/ashydv/advertising-dataset)
  Predict product sales based on advertising spending.

## Note
The most important task is to identify the **target column** correctly. Try to understand what you are predicting and whether the problem is a **classification** or a **regression** task before building the model.
