## What is Mutable in Python? 🧠

###Mutable means:
👉 You can change the value of an object after it is created, without creating a new object.

###Simple Definition (Beginner Friendly)

If a data type is mutable, you can:

* Modify it

* Add items to it

* Remove items from it

If it is immutable, you cannot change it once it’s created.

###🔁 Mutable vs Immutable (Quick Idea)

**Think of it like this:**

* Mutable → ✏️ You can edit it

* Immutable → 🔒 Read-only, no editing


## 5. Variables & Data Types

### What is a Variable?
A variable is a **named container** for storing data.

### Data Types in Python
| Data Type | Mutable | Indexed | Example |
|---------|--------|--------|--------|
| String (`str`) | ❌ No | ✅ Yes | "Python" |
| Integer (`int`) | ❌ No | ❌ No | 10 |
| Float (`float`) | ❌ No | ❌ No | 3.14 |
| Boolean (`bool`) | ❌ No | ❌ No | True |
| List (`list`) | ✅ Yes | ✅ Yes | [1,2,3] |
| Tuple (`tuple`) | ❌ No | ✅ Yes | (1,2) |
| Dictionary (`dict`) | ✅ Yes | ❌ No | {"a":1} |
| Set (`set`) | ✅ Yes | ❌ No | {1,2} |


```python
def add_numbers(num1, num2):
  return num1 + num2

# Example usage:
result = add_numbers(5, 7)
print(f"The sum is: {result}")
```
