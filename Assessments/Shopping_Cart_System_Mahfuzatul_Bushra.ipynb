```python
class ShoppingCart:

    def __init__(self, customer_name, items=None):
        self.customer_name = customer_name
        self.items = items if items is not None else []

    # 1. Add item
    def add_item(self, price):
        if price > 0:
            self.items.append(price)
            print(f"Item {price} added successfully.")
        else:
            print("Price must be a positive number.")

    # 2. Remove item
    def remove_item(self, price):
        if price in self.items:
            self.items.remove(price)
            print(f"Item {price} removed successfully.")
        else:
            print("Item not found in cart.")

    # 3. Calculate total
    def calculate_total(self):
        return sum(self.items)

    # 4. Apply discount
    def apply_discount(self):
        total = self.calculate_total()

        if total >= 3000:
            discount = total * 0.10
            final_amount = total - discount
        else:
            final_amount = total

        return final_amount

    # Display cart details
    def display_cart(self):
        total = self.calculate_total()
        final_amount = self.apply_discount()

        print("\n----- Shopping Cart Details -----")
        print(f"Customer Name: {self.customer_name}")
        print(f"Items: {self.items}")
        print(f"Total Amount: {total}")
        print(f"Final Amount (after discount): {final_amount}")
        print(f"Number of Items: {len(self.items)}")
        print("--------------------------------")

    # Clear cart
    def clear_cart(self):
        self.items.clear()
        print("Cart has been cleared successfully.")
```

```python
# Create cart
cart1 = ShoppingCart("Bushra")

# Add items
cart1.add_item(1000)
cart1.add_item(1500)
cart1.add_item(800)

# Remove item
cart1.remove_item(1500)

# Display cart
cart1.display_cart()

# Clear cart
cart1.clear_cart()

# Display again
cart1.display_cart()
```
