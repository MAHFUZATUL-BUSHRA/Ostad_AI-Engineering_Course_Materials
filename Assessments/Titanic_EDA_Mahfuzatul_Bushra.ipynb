```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
```

```python
df=pd.read_csv(url)
```

```python
survival_Rate=df['Survived'].mean()*100
```

```python
import matplotlib.pyplot as plt
```

```python
print(survival_Rate)
```

```python
sns.countplot(x='Pclass', data=df)
plt.title("Passenger Class Distribution")
plt.show()
```

```python
df['Pclass'].value_counts()
```

```python
pd.crosstab(df['Sex'],df['Survived'])
```

```python
sns.countplot(x='Survived',hue='Sex', data=df)
plt.title("Survuval by Sex")
plt.show()
```

```python
df.groupby('Pclass')['Survived'].mean()
```

```python
sns.barplot(x='Pclass',y='Survived',data=df)
plt.title("Survival Rate by Class")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(df[df['Survived']==1]['Age'], label='Survived', fill=True)
sns.kdeplot(df[df['Survived']==0]['Age'], label='Not Survived', fill=True)
plt.legend()
plt.title("Age Distribution by Survival")
plt.show()
```

```python
df.groupby('Embarked')['Survived'].mean()
```

```python
sns.barplot(x='Embarked', y='Survived', data=df)
plt.title("Survival by Embarked Port")
plt.show()
```

Conclusion:
👩‍🦰 Gender was a major factor in survival. Female passengers had a significantly higher survival rate compared to males, showing they were prioritized during rescue.
🎟️ Passenger class strongly influenced survival. First-class passengers had the highest survival rate, while third-class passengers had the lowest, indicating socio-economic advantage.
👶 Age played an important role. Children had better survival chances compared to older passengers, suggesting they were given priority during evacuation.
📊 Overall insight: Sex, passenger class, and age are the strongest predictors of survival on the Titanic dataset.
